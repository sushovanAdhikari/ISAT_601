```python
from sklearn.datasets import load_iris
from sklearn import tree
import numpy as np

iris = load_iris()
test_indexes = [0,50,100]

for i in range(len(iris.target)):
    print(f'Example {i}: Label {iris.target[i]}, Features {iris.data[i]}')
```

```python
#training data
train_target = np.delete(iris.target, test_indexes)
train_data = np.delete(iris.data, test_indexes, axis = 0)

#testing data
test_target = np.take(iris.target, test_indexes)
test_data = np.take(iris.data, test_indexes)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_target)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure()
plot_tree(clf, filled=True)
plt.title("Decision tree trained on all the iris features")
plt.show()
plt.savefig('filename.png', dpi=300)
```
